```python
import pandas as pd
import numpy as np
```

```python
df_cork = pd.read_csv('C:/Users/buckl/dev/Anna-BT-Project-Data-Analysis/bt/Swithins Day/Rainfall 15 July-24 August (Cork).csv')
df_dublin = pd.read_csv('C:/Users/buckl/dev/Anna-BT-Project-Data-Analysis/bt/Swithins Day/Rainfall 15 July-24 August (Dublin).csv')
df_galway = pd.read_csv('C:/Users/buckl/dev/Anna-BT-Project-Data-Analysis/bt/Swithins Day/Rainfall 15 July-24 August (Galway).csv')
df_donegal = pd.read_csv('C:/Users/buckl/dev/Anna-BT-Project-Data-Analysis/bt/Swithins Day/Rainfall 15 July-24 August (Donegal).csv')
```

```python
df_cork.drop('Day',axis=1,inplace=True)
df_dublin.drop('Day',axis=1,inplace=True)
df_galway.drop('Day',axis=1,inplace=True)
df_donegal.drop('Day',axis=1,inplace=True)
```

```python
swithins_total = df_cork.iloc[0]+df_dublin.iloc[0]+df_galway.iloc[0]+df_donegal.iloc[0]
```

```python
preceding_40_days_df = df_cork.iloc[1:].sum()+df_dublin.iloc[1:].sum()+df_donegal.iloc[1:].sum()+df_galway.iloc[1:].sum()
```

```python
swithins = pd.DataFrame({
    'swithins_rainfall':swithins_total,
    'next_40_days_rainfall':preceding_40_days_df
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the style for the plot
sns.set_theme(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='swithins_rainfall', y='next_40_days_rainfall', data=swithins)

# Add title and labels
plt.title('Correlation Between Rainfall on St. Swithin\'s Day and the Next 40 Days', fontsize=16)
plt.xlabel('Rainfall on St. Swithin\'s Day (mm)', fontsize=12)
plt.ylabel('Rainfall in Next 40 Days (mm)', fontsize=12)

# Show the plot
plt.show()
```

```python
corr = swithins['swithins_rainfall'].corr(swithins['next_40_days_rainfall'])



# Set the style for the plot
sns.set_theme(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.regplot(x='swithins_rainfall', y='next_40_days_rainfall', data=swithins)

plt.text(0.05, 0.95, f'Correlation: {corr:.3f}', transform=plt.gca().transAxes, fontsize=14, verticalalignment='top')

# Add title and labels
plt.title('Correlation Between Rainfall on St. Swithin\'s Day and the Next 40 Days', fontsize=16)
plt.xlabel('Rainfall on St. Swithin\'s Day (mm)', fontsize=12)
plt.ylabel('Rainfall in Next 40 Days (mm)', fontsize=12)

# Show the plot
plt.show()
```
