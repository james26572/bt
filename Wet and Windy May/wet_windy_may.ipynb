```python
rain_cork_df
```

```python
rain_cork_df = rain_cork_df.drop(['Day','2024'],axis=1).sum(axis=0)
wind_cork_df = wind_cork_df.drop(['Day','2024'],axis=1).sum(axis=0)

rain_dub_df = rain_dub_df.drop(['Day','2024'],axis=1).sum(axis=0)
wind_dub_df = wind_dub_df.drop(['Day','2024'],axis=1).sum(axis=0)

rain_don_df = rain_don_df.drop(['Day','2024'],axis=1).sum(axis=0)
wind_don_df = wind_don_df.drop(['Day','2024'],axis=1).sum(axis=0)



```

```python
rain_data = {
    'Cork': rain_cork_df,
    'Dublin': rain_dub_df,
    'Donegal': rain_don_df
}

wind_data = {
    'Cork': wind_cork_df,
    'Dublin': wind_dub_df,
    'Donegal': wind_don_df
}

rain_combined = pd.DataFrame(rain_data)
wind_combined = pd.DataFrame(wind_data)

rain_df = rain_combined.mean(axis=1)
wind_df = wind_combined.mean(axis=1)
```

```python
df = pd.DataFrame({'rain':rain_df.values,'wind':wind_df.values,'y':hay_df['tonnes of hay']})
```

```python
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y):
    mi_scores = mutual_info_regression(X, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
X = df.copy()
```

```python
y = X.pop('y')
```

```python
mi_scores = make_mi_scores(X, y)
mi_scores
```

```python

from sklearn.cluster import KMeans

# Clustering using x1, x2, and y to find subgroups in the trend
X = df[['rain', 'wind']].values

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

```

```python
sns.lmplot(x='wind',y='y',hue='cluster',data=df)
```

```python
from sklearn.linear_model import LinearRegression

# Assuming data is your dataframe and contains x1, y, and cluster results

# Choose x1 as your independent variable
X = df[['wind', 'cluster']]  # Including x1 and cluster as input features

# The dependent variable
y = df['y']

# Convert cluster column to a categorical variable (if it's not already)
#X['cluster'] = X['cluster'].astype('category')
df['cluster'] = df['cluster'].astype(int)
X['x1_cluster_interaction'] = X['wind'] * X['cluster']
X = pd.get_dummies(X, columns=['cluster'], drop_first=True)




# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Display the coefficients and intercept
print(f'Coefficients: {model.coef_}')
print(f'Intercept: {model.intercept_}')

# Predict y using the fitted model
y_pred = model.predict(X)

# Evaluate the model (e.g., using R^2)
r_squared = model.score(X, y)
print(f'R^2: {r_squared}')
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Create polynomial features (up to degree 2)
poly = PolynomialFeatures(degree=2, interaction_only=True)
X_poly = poly.fit_transform(df[['rain', 'wind']])

# Fit the linear model
model = LinearRegression()
model.fit(X_poly, df['y'])

# Evaluate the model
print(f"R^2: {model.score(X_poly, df['y'])}")
```

```python
from sklearn.tree import DecisionTreeRegressor

# Train a decision tree
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(df[['rain', 'wind']], df['y'])

# Evaluate the model
print(f"R^2: {tree_model.score(df[['rain', 'wind']], df['y'])}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train a random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(df[['rain', 'wind']], df['y'])

# Evaluate the model
print(f"R^2: {rf_model.score(df[['rain', 'wind']], df['y'])}")
```

```python
from sklearn.neural_network import MLPRegressor

# Initialize a neural network model
nn_model = MLPRegressor(hidden_layer_sizes=(100,), random_state=42)

# Train the model
nn_model.fit(df[['rain', 'wind']], df['y'])

# Evaluate the model
print(f"R^2: {nn_model.score(df[['rain', 'wind']], df['y'])}")
```
